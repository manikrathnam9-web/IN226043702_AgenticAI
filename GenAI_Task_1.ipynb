# Task 1 & 2: Replacing Hardcoded Prompts & Multi-Input Systems

```python
from langchain_core.prompts import PromptTemplate

# Task 1: Basic PromptTemplate [cite: 50]
template_1 = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} in simple terms for beginners"
)

# Task 2: Multi-Input System [cite: 54, 58]
multi_input_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone"
)

# Test Cases for Task 2 [cite: 60]
test_cases = [
    {"topic": "AI", "audience": "beginners", "tone": "friendly"},
    {"topic": "Python", "audience": "kids", "tone": "fun"},
    {"topic": "Deep Learning", "audience": "engineers", "tone": "technical"}
]

for test in test_cases:
    print(multi_input_template.format(**test))
```

# Task 3 & 4: Variations & Chat-Based Systems

```python
from langchain_core.prompts import ChatPromptTemplate, PromptTemplate

# Task 3: Prompt Variations Engine [cite: 66]
teaching_template = PromptTemplate.from_template("Teaching: Explain {topic} clearly step by step")
interview_template = PromptTemplate.from_template("Interview: Ask 3 questions about {topic}")
storytelling_template = PromptTemplate.from_template("Storytelling: Explain {topic} as a story")

topic_val = "Machine Learning"
print(teaching_template.format(topic=topic_val))
print(interview_template.format(topic=topic_val))
print(storytelling_template.format(topic=topic_val))

# Task 4: ChatPrompt Template [cite: 74]
chat_template = ChatPromptTemplate.from_messages([
    ("system", "You are a {role}."),
    ("user", "Explain {topic}")
])

print(chat_template.format_messages(role="teacher", topic="Neural Networks"))
```

# Task 5 & 6: Input Validation & Prompt Generator App

```python
from langchain_core.prompts import PromptTemplate

# Task 5: Validation Layer [cite: 89, 91]
def validate_inputs(audience, tone):
    valid_audiences = ["beginner", "intermediate", "expert"]
    valid_tones = ["formal", "casual", "fun"]

    # Assigning default values if invalid [cite: 95]
    final_audience = audience if audience in valid_audiences else "beginner"
    final_tone = tone if tone in valid_tones else "formal"

    return final_audience, final_tone

# Task 6: Prompt Generator Function [cite: 98, 102]
def generate_prompt(topic, audience, tone, style):
    val_audience, val_tone = validate_inputs(audience, tone)

    styles = {
        "teaching": "Explain {topic} for {audience} in a {tone} teaching style.",
        "interview": "Ask interview questions about {topic} for an {audience} level in a {tone} tone.",
        "storytelling": "Explain {topic} for {audience} in a {tone} storytelling style."
    }

    selected_template = PromptTemplate.from_template(styles.get(style, styles["teaching"]))
    return selected_template.format(topic=topic, audience=val_audience, tone=val_tone)

# Example Output [cite: 104]
print(generate_prompt("Neural Networks", "beginner", "fun", "storytelling"))
```
