```python
# ==========================================
# Install Required Libraries (Run this first if needed)
# ==========================================

import torch
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re

from datasets import load_dataset
from transformers import (
    AutoTokenizer,
    AutoModelForSequenceClassification,
    TrainingArguments,
    Trainer,
    EarlyStoppingCallback
)
from sklearn.metrics import accuracy_score, precision_recall_fscore_support, confusion_matrix

print("--- Step 1 & 2: Data Loading, Preprocessing, and Splitting ---")
dataset = load_dataset("imdb")
train_dataset = dataset['train'].shuffle(seed=42).select(range(2000))
test_val_dataset = dataset['test'].shuffle(seed=42).select(range(1000))

# Splitting test into validation and test sets (500 each)
val_dataset = test_val_dataset.select(range(500))
test_dataset = test_val_dataset.select(range(500, 1000))

# Preprocessing function to clean HTML tags and weird characters
def clean_text(example):
    text = example['text']
    text = re.sub(r'<.*?>', '', text) # Remove HTML
    text = re.sub(r'[^a-zA-Z\s]', '', text) # Keep only letters
    text = text.lower()
    text = re.sub(r'\s+', ' ', text).strip()
    return {'text': text}

print("Cleaning text data...")
train_dataset = train_dataset.map(clean_text)
val_dataset = val_dataset.map(clean_text)
test_dataset = test_dataset.map(clean_text)


print("\n--- Step 3: Tokenization ---")
model_name = "bert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(model_name)

def tokenize_function(examples):
    # Truncating to 128 tokens for faster training. For production, 512 is standard.
    return tokenizer(examples["text"], padding="max_length", truncation=True, max_length=128)

print("Tokenizing datasets...")
tokenized_train = train_dataset.map(tokenize_function, batched=True)
tokenized_val = val_dataset.map(tokenize_function, batched=True)
tokenized_test = test_dataset.map(tokenize_function, batched=True)


print("\n--- Step 4: Define Evaluation Metrics ---")
def compute_metrics(pred):
    labels = pred.label_ids
    preds = pred.predictions.argmax(-1)
    precision, recall, f1, _ = precision_recall_fscore_support(labels, preds, average='binary')
    acc = accuracy_score(labels, preds)
    return {
        'accuracy': acc,
        'f1': f1,
        'precision': precision,
        'recall': recall
    }

def plot_confusion_matrix(y_true, y_pred, title):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(6,4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Negative', 'Positive'], yticklabels=['Negative', 'Positive'])
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.title(title)
    plt.show()


# ==========================================
# EXPERIMENT SETUP
# ==========================================
def run_experiment(experiment_name, freeze_all_bert=False, unfreeze_last_n_layers=0):
    print(f"\n========== RUNNING: {experiment_name} ==========")

    # 1. Load fresh model
    model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=2)

    # 2. Freezing Logic
    if freeze_all_bert:
        # Freeze the entire BERT backbone
        for param in model.bert.parameters():
            param.requires_grad = False

    if unfreeze_last_n_layers > 0:
        # Unfreeze only the last N layers of the encoder
        for param in model.bert.encoder.layer[-unfreeze_last_n_layers:].parameters():
            param.requires_grad = True

    # 3. Training Arguments (Includes Bonus: LR Scheduler is default linear, plus Early Stopping logic)
    training_args = TrainingArguments(
        output_dir=f"./results_{experiment_name.replace(' ', '_')}",
        eval_strategy="epoch",      # Evaluate every epoch
        save_strategy="epoch",      # Save model every epoch (needed for early stopping)
        learning_rate=2e-5,         # As requested in task
        per_device_train_batch_size=16,
        per_device_eval_batch_size=16,
        num_train_epochs=5,         # Setting to 5, but early stopping will halt it if it stops improving
        weight_decay=0.01,
        load_best_model_at_end=True,
        metric_for_best_model="accuracy",
        logging_dir='./logs',
        logging_steps=50,
        report_to="none"            # Prevents weights&biases prompt
    )

    # 4. Initialize Trainer (Includes Bonus: Early Stopping)
    trainer = Trainer(
        model=model,
        args=training_args,
        train_dataset=tokenized_train,
        eval_dataset=tokenized_val,
        compute_metrics=compute_metrics,
        callbacks=[EarlyStoppingCallback(early_stopping_patience=2)] # Stops if no improvement after 2 epochs
    )

    # 5. Train
    print("Training started...")
    trainer.train()

    # 6. Evaluate on Test Set
    print("\nEvaluating on Test Set...")
    predictions = trainer.predict(tokenized_test)
    metrics = predictions.metrics

    print(f"Test Metrics for {experiment_name}:")
    print(metrics)

    # 7. Plot Confusion Matrix
    y_true = tokenized_test['label']
    y_pred = predictions.predictions.argmax(-1)
    plot_confusion_matrix(y_true, y_pred, f"Confusion Matrix: {experiment_name}")

    return metrics

# ==========================================
# Step 6 & 7: Execution and Experiments
# ==========================================
# Experiment 1: Freeze BERT entirely, train only the classification head
metrics_exp1 = run_experiment("Exp 1 - Frozen BERT Layers", freeze_all_bert=True, unfreeze_last_n_layers=0)

# Experiment 2: Fine-tune the last 2 layers of BERT + classification head
metrics_exp2 = run_experiment("Exp 2 - Fine-Tuning Last 2 Layers", freeze_all_bert=True, unfreeze_last_n_layers=2)
```
